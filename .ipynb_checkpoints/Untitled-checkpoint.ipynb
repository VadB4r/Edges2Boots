# Проект DLS

## Генерация обуви из контуров с помощью pix2pix

В своем проекте я создал бота, который генерирует обувь из контура при помощи pix2pix. Конкретно в этом файле описан процесс обучения моделей: Генератора и Дискриминатора.

```python
import os
from torch.utils.data import DataLoader
from torchvision.datasets import ImageFolder
import torchvision.transforms as tt
import torch
import torch.nn as nn
import cv2
from tqdm.notebook import tqdm
from torchvision.utils import save_image
from torchvision.utils import make_grid
from models.generator import Generator
from models.discriminator import Discriminator
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set(style='darkgrid', font_scale=1.2)
```

Записываем постоянные и пишем функции, которые создают DataLoader, денормируют изображения и переносят данные на устройство.

```python
size = (256, 512)
image_size = 256
patch_size = 16
batch_size = 16
latent_size = 64
size = 3 * 256 * 256
lamb = 100
lr = 0.0002
beta_1 = 0.5
beta_2 = 0.999
epochs = 1
stats = (0.5, 0.5, 0.5), (0.5, 0.5, 0.5)
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
```

```python
def get_dataloader(batch_size, train=True):
    path = 'archive/train' if train else 'archive/val'
    ds = ImageFolder(path, transform=tt.Compose([
        tt.ToTensor(),
        tt.Normalize(*stats)]))
    dl = DataLoader(ds, batch_size, shuffle=True, num_workers=2, pin_memory=True)
    return dl
```

```python
def denorm(img_tensors):
    return img_tensors * stats[1][0] + stats[0][0]
```

```python
def to_device(data, device):
    """Move tensor(s) to chosen device"""
    if isinstance(data, (list,tuple)):
        return [to_device(x, device) for x in data]
    return data.to(device, non_blocking=True)

class DeviceDataLoader():
    """Wrap a dataloader to move data to a device"""
    def __init__(self, dl, device):
        self.dl = dl
        self.device = device
        
    def __iter__(self):
        """Yield a batch of data after moving it to device"""
        for b in self.dl: 
            yield to_device(b, self.device)

    def __len__(self):
        """Number of batches"""
        return len(self.dl)
```

```python
train_dl = get_dataloader(batch_size)
train_dl = DeviceDataLoader(train_dl, device)
```

Создадим дискриминатор и генератор, которые мы импортировали и файлов с соответствующими классами.

```python
discriminator = Discriminator().to(device)
generator = Generator().to(device)
```

Здесь для удобства засовываем их в словарь и так же создаем словарь функций, с помощью которых будем строить функцию потерь.

```python
model = {
    "discriminator": discriminator,
    "generator": generator
}

criterion = {
    "BCE": nn.BCELoss(),
    "L1": nn.L1Loss()
}
```

Описываем функцию обучения моделей.

```python
def fit(model, criterion, epochs, lr, beta_1, beta_2, start_idx=1, add_learn=False, path_dis='', path_gen=''):
    if add_learn:
        model["generator"].load_state_dict(torch.load(path_gen))
        model["discriminator"].load_state_dict(torch.load(path_dis))
    model["discriminator"].train()
    model["generator"].train()
    torch.cuda.empty_cache()
    
    # Losses & scores
    losses_g = []
    losses_d = []
    
    # Create optimizers
    optimizer = {
        "discriminator": torch.optim.Adam(model["discriminator"].parameters(), 
                                          lr=lr, betas=(beta_1, beta_2)),
        "generator": torch.optim.Adam(model["generator"].parameters(),
                                      lr=lr, betas=(beta_1, beta_2))
    }
    
    for epoch in range(epochs):
        loss_d_per_epoch = []
        loss_g_per_epoch = []
        for real_images, _ in tqdm(train_dl):
            
            #Inputs
            real_A = real_images[:, :, :, :256]
            real_B = real_images[:, :, :, 256:]
            
            
            #Ground truths
            valid = torch.ones([len(real_A), 1, patch_size, patch_size], requires_grad=False, device=device)
            fake = torch.zeros([len(real_A), 1, patch_size, patch_size], requires_grad=False, device=device)
            
            #Train Gen
            
            optimizer["generator"].zero_grad()
            
            #Gen Loss
            fake_B = model["generator"](real_A)
            pred_fake = model['discriminator'](torch.cat((fake_B, real_A), 1))
            loss_GAN = criterion["BCE"](pred_fake, valid)
            
            loss_L1 = criterion["L1"](fake_B, real_B)
            
            loss_G = loss_GAN + lamb * loss_L1
            
            loss_G.backward(retain_graph=True)
            
            optimizer["generator"].step()
            
            loss_g_per_epoch.append(loss_G.item())
            
            #Train Disc
            
            optimizer["discriminator"].zero_grad()
            
            #Real Loss
            pred_real = model['discriminator'](torch.cat((real_B, real_A), 1))
            loss_real = criterion["BCE"](pred_real, valid)
            
            #Fake Loss
            pred_fake = model['discriminator'](torch.cat((fake_B, real_A), 1))
            loss_fake = criterion["BCE"](pred_fake, fake)
            
            loss_D = 0.5 * (loss_real + loss_fake)

            loss_D.backward()
            
            optimizer["discriminator"].step()
            
            loss_d_per_epoch.append(loss_D.item())
            
        # Record losses & scores
        losses_g.append(np.mean(loss_g_per_epoch))
        losses_d.append(np.mean(loss_d_per_epoch))
        
        # Log losses & scores (last batch)
        print("Epoch [{}/{}], loss_g: {:.4f}, loss_d: {:.4f}".format(
            epoch+1, epochs, 
            losses_g[-1], losses_d[-1]))
    
    return losses_g, losses_d
```

Обучение.

```python
history = fit(model, criterion, epochs, lr, beta_1, beta_2, 
              add_learn=True, path_dis='weights/dis_50k_1', path_gen='weights/gen_50k_1')
```

Финальная версия обучалась всего две эпохи, которые длились 26 часов в общем.

```python
losses_g, losses_d = history
```

```python
plt.figure(figsize=(15, 6))
plt.plot(losses_d, '-')
plt.plot(losses_g, '-')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['Discriminator', 'Generator'])
plt.title('Losses');
```

```python
model["generator"].eval()
for val_images, _ in tqdm(val_dl):
    real_A = val_images[:3, :, :, :256]
    real_B = val_images[:3, :, :, 256:]
    fake_B = model["generator"](real_A)
    fig, ax = plt.subplots(figsize=(batch_size, 3))
    ax.set_xticks([]); ax.set_yticks([])
    ax.imshow(make_grid(denorm(torch.cat((real_A, real_B, fake_B), 0).cpu().detach()), nrow=batch_size).permute(1, 2, 0))
```
